```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Open_powerlifting.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
#drop

df.drop("Place", axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
#mean

df.fillna(df["Deadlift4Kg"].mean(), inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Place']
```

```python
#mode

df.fillna(df["Bench4Kg"].mode()[0], inplace=True)
```

```python
df.head()
```

```python
missing_col = df.columns[df.isnull().sum()>0]
for col in missing_col:
    if df[col].dtype=="object":
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)
```
