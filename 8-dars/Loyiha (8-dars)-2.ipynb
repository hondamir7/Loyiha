```python
#Kutubxona yuklab olish

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Ma'lumotlarni yuklab olish

df=pd.read_csv('Corporate Stress.csv')
```

```python
df.info()
```

```python
#Tushib qolgan qiymatlarni ko'rish

df.isnull().sum()
```

```python
#'Health_Issues' ustunini o'chirish

df=df.drop(columns=['Health_Issues'])
```

```python
#Missing valueslarni to'ldirib olish

missing_col=df.isnull().sum()>0
for col in df.columns[missing_col]:
    if df[col].dtype=='object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#Mavjud boolean ustunlarni raqamlashtirib olish

df=df.astype({col:'int' for col in df.select_dtypes(include=['bool']).columns})
```

```python
df.info()
```

```python
#Label Encodingni chaqirib olamiz

from sklearn.preprocessing import LabelEncoder
```

```python
#Object tipidagi ustunlarni LabelEncoding qilamiz. Qolgan ustunlarni OneHotEncoding qilamiz

for col in df.select_dtypes(include=['object']).columns:
    cardinality=df[col].nunique()
    if cardinality<=5:
        le=LabelEncoder()
        df[col]=le.fit_transform(df[col])
    else:
        df=pd.get_dummies(df, columns=[col], dtype=int, drop_first=True)
```

```python
df.info()
```

```python
#Min-max Scalingni tortib, uni belgilab olamiz

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
#Datani Min-max Scaling qilamiz

df=scaler.fit_transform(df)
```

```python
df=pd.DataFrame(df)
```
